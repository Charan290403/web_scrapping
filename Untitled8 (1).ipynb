# ***Web Scraping using Beatufullsoup***

# ***Importing the modules***

```python
from bs4 import BeautifulSoup
import requests
```

***Adding the URL***

```python
url='https://companiesmarketcap.com/usa/largest-american-companies-by-revenue/'
page=requests.get(url)
soup=BeautifulSoup(page.text,'html')
```

***Print***

```python
print(soup)
```

# ***Find the table***

```python
soup.find_all('table')
```

```python
soup.find('table',class_='default-table table marketcap-table dataTable')
```

# ***find specific table***

```python
table=soup.find_all('table')[0]
print(table)
```

# ***find the headers of the table***

```python
soup.find_all('th')
```

# ***find the word title***

```python
word_titles=soup.find_all('th')
word_titles
```

# ***print the titles of the table***

```python
word_table_title=[word.text for word in word_titles]
print(word_table_title[1:])
```

# ***add them in dataframe***

```python
import pandas as pd
df=pd.DataFrame(columns=word_table_title[1:])
df
```

```python
column_data=table.find_all('tr')
print(len(column_data))
```

# ***add data to df***

```python
indicies=[0,18,76]
for row in column_data:
  row_data=row.find_all('td')
  raw_data=[data.text.strip() for data in row_data]
  row_data=[data for idx,data in enumerate(raw_data) if idx not in indicies]
  #print(len(row_data))

  length=len(df)
  if len(row_data)==7:
    df.loc[length]=row_data
  else:
    continue
df
```

# ***remove the unwanted column***

```python
df=df.drop(columns=['Price (30 days)'])
df
```
